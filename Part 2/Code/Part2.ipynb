## INFLUENZA CLINICAL HISTOGRAM

```r
clinic_health <- read.csv("clinic.csv") # Reading the Clinic health csv
total_a <- clinic_health$TOTAL.A #Read columns Total A
total_b <- clinic_health$TOTAL.B #Read columns Total B

x <- paste(clinic_health$YEAR,clinic_health$WEEK)
data <- cbind(total_b,total_a) #Combining column wise to form a df
colnames(data) <- c("B","A")
rownames(data) <- c(x)

bar <- barplot(as.matrix(t(data)),xlim = c(1.8,40),las=2,ylim = c(0,14000), col=(c("green","yellow")) #bar plotting
,main = "Influenza Positive Tests Reported to 
CDC by U.S. Clinical Laboratories, National Summary, 2018-2019 Season")

lines(x=bar,y=clinic_health$PERCENT.POSITIVE*400,lwd=2) #plotting different types of lines
lines(x=bar,y=clinic_health$PERCENT.A*400,lwd=2,lty=2,col="blue")
lines(x=bar,y=clinic_health$PERCENT.B*400,lwd=2,lty=3,col="red")

axis(side = 4, at = seq(0,14000,by=2000), labels= seq(0,35,by=5))
legend(27, 13000, legend=c("A", "B","Percent Positive","% Positive Flu A","% Positive Flu B"), #legend
       col=c("yellow", "green","black","blue","red"), lty=c(1,1,1,2,3), lwd=c(8,8,3,2,3),cex=1.0)

#Analysis - 
#The graph shows the number of people tested positive for both flu A and flu B. The black line shows the total % of people tested
#positive for the flow. The % percent positive for Fu A and Flu B are shown in blue and red dashed lines.
```

## INFLUENZA LINEAGE HISTOGRAM

```r
options(width = 200) 

public_health <- read.csv("public.csv") # Reading the Public health csv

Subtyping_A <- public_health$A..Subtyping.not.Performed.
H1N1_A <- public_health$A..2009.H1N1.
H3N2_A <- public_health$A..H3.
H3N2 <- public_health$H3N2v
Lineage_B <- public_health$B
Victoria_B <- public_health$BVic
Yamagata_B <- public_health$BYam

x <- paste(public_health$YEAR,public_health$WEEK)
data <- cbind(Yamagata_B, Victoria_B, Lineage_B, H3N2, H3N2_A, H1N1_A, Subtyping_A)
colnames(data) <- c("A","B","C","D","E","F","G")
rownames(data) <- c(x)

bar <- barplot(as.matrix(t(data)),xlab="Week",ylab = "Number of Positive Specimens",xlim = c(1.6,40),las=2,ylim = c(0,3000), col=(c("Light Green", "Green", "Dark Green", "Blue", "Red", "Orange", "Yellow"))
,main = "Influenza Positive Tests Reported to 
CDC by U.S. Public Health Laboratories, National Summary, 2018-2019 Season")

legend(26, 2800, legend=c("A (Subtyping not performed)", "A (H1N1)pdm09","A (H3N2)","H3N2v","B (Lineage not performed)","B (Victoria Lineage)","B (Yamagata Lineage)"),
       col=c("yellow", "Orange","Red","blue","Dark Green","Green","Light Green"), lty=c(1,1,1,1,1,1,1), lwd=c(5),cex=0.7)

#Analysis
#The graph shows the number of people from different lineage who were tested positive for flu. 
```

## INFLUENZA ASSOCIATED PEDIATRIC DEATH HISTOGRAM

```r
weekly <- read.csv("Weekly.csv") # Reading the Clinic health csv

prev_deaths <- weekly$PREVIOUS.WEEK.DEATHS
curr_deaths <- weekly$CURRENT.WEEK.DEATHS
data <- cbind(prev_deaths,curr_deaths)
colnames(data) <- c("B","A")
rownames(data) <- weekly$WEEK.NUMBER


bar <- barplot(as.matrix(t(data)),xlab = "Week of Death",ylab = "Number of Deaths",xlim = c(8,230),ylim = c(0,30), col=(c("Dark green","Cyan"))
              ,main = "Number of Influenza-Associated Pediatric 
Deaths by Week of Death:2015-2016 Season to present")

legend(70, 29, legend=c("Deaths Reported Previous Week", "Deaths Reported Current Week"),
       col=c("Dark green","red"), lty=c(1,1), lwd=c(8,8),cex=1)

legend(1, 25, legend=c("      2015-2016
Number of Deaths
   Reported = 94"),bty="n",cex=0.8)

legend(60, 25, legend=c("      2016-2017 
Number of Deaths
  Reported = 110"),bty="n",cex=0.8)

legend(120, 25, legend=c("      2017-2018 
Number of Deaths
  Reported = 185"),bty="n",cex=0.8)

legend(175, 25, legend=c("      2018-2019 
Number of Deaths
   Reported = 34"),bty="n",cex=0.8)


#Analysis
#The graph shows the number of pediatric deaths due to Influenza from year 2015 to current week.
```

## PNEUMONIA AND INFLUENZA MORTALITY CURVE GRAPH

```r
weekly <- read.csv("nation.csv") # Reading the Clinic health csv

baseline <- weekly$BASELINE
threshold <- weekly$THRESHOLD

x <- paste(weekly$WEEK)#,weekly$SEASON)
data <- cbind(weekly$PERCENT.P.I)
colnames(data) <- c("A")
rownames(data) <- c(x)

plot(data,ylim = c(4,12),type = "l",col = "red",xlab = "MMWR Week",ylab = "% of all deaths due to P & I", xaxt="n",
     main = "Pneumonia and Influenza Mortality from the National 
Center for Health Statistice Mortality Surveillance System
     Data through the week ending February 23, 2019, as of March 7, 2019")

axis(1,at=1:229,labels = c(x))

lines(threshold,type = "l")
lines(baseline, type = "l")

legend(35, 8.2, legend=c("Epidemic 
Threshold"),bty="n",cex=0.8)
legend(65, 5.8, legend=c("Seasonal
 Baseline"),bty="n",cex=0.8)
legend(1, 4.5, legend=c("2014"),bty="n",cex=0.8)
legend(30, 4.5, legend=c("2015"),bty="n",cex=0.8)
legend(90, 4.5, legend=c("2016"),bty="n",cex=0.8)
legend(140,4.5, legend=c("2017"),bty="n",cex=0.8)
legend(200, 4.5, legend=c("2018"),bty="n",cex=0.8)


#Analysis
#The graph shows the pneumonia and Influenza Mortality rates to the current week. 
```

## INFLUENZA HEAT MAP

```r
require("devtools")
devtools::install_github("UrbanInstitute/urbnmapr")
library(ggplot2)
library(ggmap)
library(maps)
library(mapdata)
library(devtools)
library(dplyr)
library(urbnmapr)

map_data <- read.csv("map.csv")

data_to_plot <- data.frame(map_data$STATENAME,map_data$ACTIVITY.LEVEL.LABEL)
colnames(data_to_plot) <- c("state_name","level")
data_to_plot <- slice(data_to_plot,1:53)


data_to_plot %>% 
  left_join(states, by = "state_name") %>% 
  ggplot(mapping = aes(long, lat, group = group, fill = level)) +geom_polygon(color = "#ffffff", size =0.3) + coord_fixed(1.3)


#Analysis
#The graph or a heatmap shows the US map and highlight each state with its corresponding flu activity level.
```
