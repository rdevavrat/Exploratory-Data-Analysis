```r
fb1<-read.csv(file.choose())
aapl1<-read.csv(file.choose())
par(bg="cornsilk")
plot(aapl1$Adj.Close, col="blue", type="o", ylim=c(120,250), xlab="Days", ylab="Price" )
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)
hist(aapl1$Adj.Close, col=rainbow(8))

#Analysis - 
#The plot is comparison of closing prices of the stocks of apple and facebook over a period of approximately 60 days.
#The range of the closing prices of both the companies are within 170 and 190. (Except there is an outlier in Apple s data)

#The histogram shows the number of occurences of the closing price of the apple stock prices.
#Over the period of 60 days, apple s closing price is at 175 for most of the days.
```
